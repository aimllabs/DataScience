### Checkpoint 1: Data Cleaning 1

```python
#importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# loading companies file
df_companies = pd.read_csv('companies.txt',delimiter="\t", encoding = 'palmos')
```

```python
# Removing special character in companies file
df_companies.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
```

```python
df_companies.head(5)
```

```python
# loading rounds2 file
df_rounds2 = pd.read_csv("rounds2.csv",encoding = "unicode_escape")
```

```python
# Removing special character in companies file
df_rounds2.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
```

```python
df_rounds2.head(5)
```

```python
# Converting each company name to lower case to comapre the unique value.
df_rounds2['company_permalink'] = df_rounds2['company_permalink'].apply(lambda x: x.lower())
```

```python
# Table 1.1.1 Q. How many unique companies are present in rounds2?
df_rounds2['company_permalink'].nunique()
```

```python
# Converting each company name to lower case to comapre the unique value.
df_companies['permalink'] = df_companies['permalink'].apply(lambda x: x.lower())
```

```python
# Table 1.1.2 Q. How many unique companies are present in companies?
df_companies['permalink'].nunique()
```

```python
# Table 1.1.3 Q. In the companies data frame, which column can be used as the unique key for each company? Write the name of the column.	
#Ans will be-- permalink
df_companies.head(5)
```

```python
# Table 1.1.4 Q. Are there any companies in the rounds2 file which are not  present in companies ? Answer Y/N.
# Ans is --N
set(df_companies['permalink'].unique()).difference(set(df_rounds2['company_permalink'].unique()))
```

```python
#Inorder to merge, first rename the column to make it inner merge
df_companies.rename(columns = {'permalink':'company_permalink'}, inplace = True)
```

```python
df_companies.head(5)
```

```python
# Merge the two data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame.
# Name the merged frame master_frame.
master_frame = pd.merge(df_companies,df_rounds2,how='inner', on='company_permalink')
```

```python
master_frame.head(5)
```

```python
# Table 1.1.5 Q.How many observations are present in master_frame?
# Ans is 114949 entries
master_frame.info()
```

```python
# Before going to next section, clean the data
master_frame.isnull().sum(axis=0)
```

```python
# summing up the missing values (column-wise)
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# removing the unecessary columns which will be not the part of analysis

master_frame = master_frame.drop(['funded_at', 'funding_round_code','funding_round_permalink', 'founded_at',
                                  'homepage_url','state_code','city', 'region'], axis = 1)
```

```python
 #Removed all the closed status companies
master_frame =master_frame[master_frame['status']!='closed']
```

```python
master_frame.head(5)
```

```python
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# Cleaning the row data. As raised_amount_usd contains a lot of missing data and imputing 17% values of a variable will introduce heavy bias
# will delete NAN value in raised_amount_usd
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
```

```python
#Removing the rows after cleaning the null values for country_code
master_frame=master_frame[~master_frame.country_code.isnull()]
```

```python
#Percentage of number of rows having > 5 missing values
100*(len(master_frame[master_frame.isnull().sum(axis=1) > 5].index) / len(master_frame.index))
```

```python
# Percentage of NAN values in data
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
master_frame.head(5)
```

```python
#Extract the primary sector of each category list from the category_list column
top9['primary sector'] =top9.category_list.str.split('|').str[0]
```

```python
#Loading mapping file
mapping = pd.read_csv("mapping.csv",encoding = "unicode_escape")
```

```python
mapping.head(5)
```

```python
# Enhance mapping df with melt function
mapping = pd.melt(mapping, id_vars =['category_list'], value_vars =['Manufacturing','Automotive & Sports','Cleantech / Semiconductors','Entertainment','Health','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising']) 
```

```python
mapping.head(5)
```

```python
# need that rows only where mapping value =1
mapping = mapping[mapping.value == 1]
```

```python
mapping.head(5)
```

```python
# Now lets delete value column and rename variable column to main_sector
mapping = mapping.drop('value', axis = 1)
mapping = mapping.rename(columns = {"variable":"main_sector"})
```

```python
#Renaming category_list as well to 'primary sector' so that we can do innner join
mapping = mapping.rename(columns = {"category_list":"primary sector"})
```

```python
mapping.head(5)
```

```python
top9.head(5)
```

```python
# Converting each primary sector to lower case to comapre the unique value.
mapping['primary sector'] = mapping['primary sector'].apply(lambda x: x.lower())
```

```python
mapping.head(5)
```

```python
# Converting each primary sector to lower case to comapre the unique value.
top9['primary sector'] = top9['primary sector'].str.lower()
```

```python
top9.head(5)
```

```python
#Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also considered one of the main sectors

master_data = pd.merge(top9,mapping,on='primary sector',how='inner')
```


### Checkpoint 5: Sector Analysis 2


```python
#Create three separate data frames D1, D2 and D3 for each of the three countries containing the observations of funding type FT falling within the 5-15 million USD range. 

D1 = master_data[(master_data['country_code'] == 'USA') & (master_data['raised_amount_usd'] >= 5000000) & (master_data['raised_amount_usd'] <= 15000000)]
```

```python
D2 = master_data[(master_data['country_code'] == 'GBR') & (master_data['raised_amount_usd'] >= 5000000) & (master_data['raised_amount_usd'] <= 15000000)]
```

```python
# All the columns of the master_frame along with the primary sector and the main sector
D2.head(10)
```

```python
D2 = D2.assign(Investment_Count  = D2.groupby('main_sector')['main_sector'].transform('count'))
```

```python
D3 = D3.assign(Investment_Count  = D3.groupby('main_sector')['main_sector'].transform('count'))
```

```python
D2.head(5)
```

```python
#The total amount invested in each main sector in a separate column
D1 = D1.assign(Investment_Amount = D1.groupby('main_sector')['raised_amount_usd'].transform('sum'))
```

```python
D2 = D2.assign(Investment_Amount = D2.groupby('main_sector')['raised_amount_usd'].transform('sum'))
```

```python
D3 = D3.assign(Investment_Amount = D3.groupby('main_sector')['raised_amount_usd'].transform('sum'))
```

#### 1. Create a plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type


```python

# Importing required libraries
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df= master_frame[master_frame['funding_round_type'].isin(['venture','seed','private_equity'])]
```

```python
#df is the merged data of companies and rounds2
df.head(2)
```

```python
# A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type
plt.figure(figsize=(10,5),dpi=80, facecolor='w', edgecolor='k')
p1= sns.barplot(y='funding_round_type',x='raised_amount_usd',data=df,estimator=np.mean)
p1.set(xlabel='Total Investment', ylabel='Funding Type')
p1.set_title('Funding Analysis',fontsize =15)
plt.show()
```

#### 2.Create a plot showing the top 9 countries against the total amount of investments of funding type FT. This should make the top 3 countries (Country 1, Country 2, and Country 3) very clear.

```python
plt.figure(figsize=(10,5),dpi=80, facecolor='w', edgecolor='k')
p2=sns.barplot(x='country_code', y='raised_amount_usd',palette="rocket",data=top9,estimator=np.sum)
p2.set_yscale('log')
p2.set(xlabel='Top 9 Countries', ylabel='Total Investment')
p2.set_title('Country Analysis',fontsize =15)
plt.show()
```
